# Euclidean algorithm for finding GCD

GCD: greatest common divisor

```python
import logging
logging.basicConfig(level=logging.DEBUG)
logger = logging.getLogger("gcd")

def gcd(a, b):
    logger.debug(f"{a = }, {b = }")    
    while b:
        a, b = b, a % b
        logger.debug(f"{a = }, {b = }")
    return a


def lcd(a, b):
    return abs(a * b) / gcd(a, b)
```

```python
# Try it
gcd(48, 18)
```

```python
# Order does not matter
gcd(18, 48)
```

```python
lcd(18, 24)
```

```python
lcd(24, 36)
```
