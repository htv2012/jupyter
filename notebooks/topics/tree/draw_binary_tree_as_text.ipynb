# Notes
1. At the lowest level, each node occupies at least 5 columns
2. There should be at least 1 column between the nodes
3. Let an empty space to denode null
4. Assume each label takes 3 columns
   - The distant between nodes at the lowest level is 3 columns
   - The distant between nodes at the next level is 9 columns
   - The distant between nodes at next level is 45 columns

````python
import io
from typing import Optional
from tree import TreeNode, max_depth

def bfs(root: Optional[TreeNode]):
    if root is None:
        return []

    depth = max_depth(root)
    nodes = [[] for _ in range(depth)]
    que = [(root, 0)]
    while que:
        node, level = que.pop(0)
        nodes[level].append(" " if node is None else str(node.val))
        if level < (depth -1):
            que.append((node.left, level + 1))
            que.append((node.right, level + 1))

    return nodes

def tree2str(root: Optional[TreeNode]) -> str:
    if root is None:
        return "None"

    nodes = bfs(root)
    buf = io.StringIO()

    # hard code for now
    buf.write(" ".join(["  ┌──┴──┐  "] * (len(nodes[4]) // 2)))
    buf.write("\n")
    buf.write(" ".join(x.center(5) for x in nodes[4]))
    buf.write("\n")
    return buf.getvalue()
````
