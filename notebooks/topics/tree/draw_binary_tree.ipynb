# Draw Binary Tree

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.DiGraph()
g.add_edge("1", "2")
g.add_edge("1", "3")
g.add_edge("2", "4")
g.add_edge("2", "5")
```

```python
pos = nx.nx_agraph.graphviz_layout(g, prog="dot")
pos
```

```python
pic = nx.draw(g, pos, with_labels=True, arrows=True, node_size=1000)
```

```python
plt.savefig("/tmp/tree.png", dpi=200, bbox_inches="tight")
```
