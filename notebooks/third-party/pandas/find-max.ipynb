# Find minimum

```python
import contextlib
import pandas as pd

df1 = pd.DataFrame(
    [
        [0.0, 1],
        [1.0, 1.5],
        [2.0, 1.7],
        [3.0, 1.9],
    ],
    columns=["foo", "bar"],
)


def find_rows(df: pd.DataFrame, expression: str, target):
    expression = expression.replace("target", "@target")
    return df.query(expression)

def find_max_bar(df: pd.DataFrame, snr: float, margin: int):
    target = snr - margin
    with contextlib.suppress(IndexError):
        rows = find_rows(df, "target > foo", target)
        return rows["foo"].iloc[-1]
    return df["bar"].iloc[0]
```
