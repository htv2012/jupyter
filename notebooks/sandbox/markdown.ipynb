```python
import dataclasses
import shlex
from itertools import takewhile

def take(tokens, terminator, separator=' '):
    return separator.join(takewhile(lambda ch: ch != terminator, tokens))
    
@dataclasses.dataclass
class Heading:
    level: int
    text: str

    def __str__(self):
        return f"{'#' * self.level} {self.text}"

    @classmethod
    def from_tokens(cls, tokens):
        heading = next(tokens)
        print(f"{heading=}")
        text = take(tokens, terminator='\n')
        obj = cls(len(heading), text)
        return obj


@dataclasses.dataclass
class Link:
    text: str
    url: str

    def __str__(self):
        return f"[{self.text}]({self.url})"

    @classmethod
    def from_tokens(cls, tokens):
        text = take(tokens, terminator=']')
        assert next(token) == "("
        url = take(tokens, terminator=')', separator='')
        obj = cls(text, url)
        return obj


def parse(text: str):
    tokens = shlex.shlex(sample)
    tokens.commenters = ""
    tokens.whitespace = tokens.whitespace.replace('\n', '')
    tokens.wordchars += "#"

    bucket = []
    for token in tokens:
        if token.startswith("#"):
            tokens.push_token(token)
            yied Heading.from_tokens(tokens)
        elif token == "[":
            yield Link.from_tokenspush_token(tokens)
            
```
