```python
import logging


logging.basicConfig(level=logging.DEBUG)
logger = logging.getLogger()

class IsInt:
    def __init__(self, lower: int = None, upper: int = None):
        self.lower = lower
        self.upper = upper
        
    def __eq__(self, val: any):
        if not isinstance(val, int):
            logger.debug(f"Expect int, but got {val!r}")
            return False

        if self.lower is not None and val < self.lower:
            logger.debug(f"Expect {val:,} > {self.lower:,}")
            return False

        if self.upper is not None and val >= self.upper:
            logger.debug(f"Expect {val:,} < {self.upper:,}")
            return False
        return True
```

```python
assert {"uid": 501} == {"uid": IsInt(lower=501, upper=600)}, "failed spectacularly"
```

```python
uid = 1121
assert {"uid": uid} == {"uid": IsInt(upper=999)}
```

```python
assert {"uid": uid} == {"uid": IsInt(lower=10_000)}
```
