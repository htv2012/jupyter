# Leetcode 168 - Excel Sheet Column Title

Given an integer `columnNumber`, return _its corresponding column title as it
appears in an Excel sheet_.

For example:

    
    
    A -> 1
    B -> 2
    C -> 3
    ...
    Z -> 26
    AA -> 27
    AB -> 28 
    ...
    



**Example 1:**

    
    
    **Input:** columnNumber = 1
    **Output:** "A"
    

**Example 2:**

    
    
    **Input:** columnNumber = 28
    **Output:** "AB"
    

**Example 3:**

    
    
    **Input:** columnNumber = 701
    **Output:** "ZY"
    



**Constraints:**

  * `1 <= columnNumber <= 2^31 - 1`



```python
class Solution:
    def convertToTitle(self, columnNumber: int) -> str:
        print(f"{columnNumber=}")
        letter = dict(zip(range(26), string.ascii_uppercase))
        stack = []
        while columnNumber > 0:
            columnNumber -= 1
            print(f"{columnNumber=}  # subtract 1 for zero-base look up")
            
            columnNumber, remainder = divmod(columnNumber, 26)
            stack.append(letter[remainder])
            print(f"{columnNumber=}, {remainder=}, {stack=}")
        out = "".join(reversed(stack))
        print(f"{out=}")
        return out
```

```python
sol = Solution()
for i in range(1, 5):
    for j in range(3):
        n = (26 * i) + j
        out = sol.convertToTitle(n)
        print(f"{n} -> {out}")
        print()
```

```python
sol.convertToTitle(701)
```

```python
import string

def title_to_number(title: str) -> int:
    title = title.upper()
    num = dict(zip(string.ascii_uppercase, range(26)))
    out = 0
    for ch in title:
        out = (out * 26) + num[ch] + 1
    return out
```

```python
title_to_number('CHICKEN')
```

```python
for title in ['HONDA', 'TOYOTA']:
    number = title_to_number(title)
    print((number, title))
```
