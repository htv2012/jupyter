# Leetcode 725 Split Linked List in Parts

```python
import itertools


for count in range(16):
    k = 4
    sub_len, extras = divmod(count, k)
    li = [sub_len + 1 if i < extras else sub_len for i in range(k)]
    # print(f"{count=}, {k=}, {sub_len=}, {extras=}, {li=}")
    acc = list(itertools.accumulate(li))
    print(count, li, acc)
```

```python
def divide(count, k):
    quot, rem = divmod(count, k)
    index = 0
    lens = [quot + 1 if i < rem else quot for i in range(k)]
    print(f"{lens=}")
    # return list(itertools.accumulate(lens))
    for ln in lens:
        index += ln
        yield index
```

```python
li = list(range(37))
k = 5
last = 0
for i in divide(len(li), k):
    print(i, li[last:i])
    last = i + 1
```

```python
def divide(n: int, k: int):
    quot, rem = divmod(n, k)
    for i in range(k):
        yield quot + 1 if i < rem else quot
```
