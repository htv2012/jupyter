# Leetcode 165. Compare Version Numbers

```python
import itertools
```

```python
v1 = "15.5"
v2 = "15.05.1"
```

```python
t1 = [int(x) for x in v1.split(".")]
t2 = [int(x) for x in v2.split(".")]
```

```python
t1
```

```python
t2
```

```python
class Solution:
    def compareVersion(self, version1: str, version2: str) -> int:
        t1 = [int(x) for x in version1.split(".")]
        t2 = [int(x) for x in version2.split(".")]
        
        for a, b in itertools.zip_longest(t1, t2, fillvalue=0):
            if a < b:
                return -1
            elif a > b:
                return 1
        return 0
```

```python
sol = Solution()
fut = sol.compareVersion
```

```python
fut("15.09", "15.9")
```

```python
fut("15.9", "15.9.1")
```

```python
fut("15.09", "15.9.0")
```
