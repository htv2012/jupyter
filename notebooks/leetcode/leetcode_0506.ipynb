```python
def find_relative_ranks(score: List[int]) -> List[int]:
    # rank is a list of (score, index, ranking)
    rank = sorted(([e, i, None] for i, e in enumerate(score)), reverse=True)
    
    for i, t in enumerate(rank, 1):
        t[2] = str(i)
    rank[0][2] = "Gold Metal"
    rank[1][2] = "Silver Metal"
    rank[2][2] = "Bronze Metal"
    
    rank.sort(key=lambda t: t[1])
    return [r for _, _, r in rank]
```

```python
find_relative_ranks(score)
```

```python
find_relative_ranks([5,4,3,2,1])
```
