# Leetcode 48 - Rotating Matrix

```python
def mprint(m):
    for row in m:
        print(" ".join(f"{c:>2}" for c in row))
```

```python
def move(mat, row, col, value, size):
    new_row = col
    new_col = size - row
    hold = mat[new_row][new_col]
    print(f"move {value} from {[row, col]} to {[new_row, new_col]}, hold {hold}")
    mat[new_row][new_col] = value
    return new_row, new_col, hold    
```

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
expected = [[7,4,1],[8,5,2],[9,6,3]]
matrix = [[5,1,9,11],[2,4,8,10],[13,3,6,7],[15,14,12,16]]
```

```python
size = len(matrix[0]) - 1
moved = set()
op_count = 0
skip_count = 0

mprint(matrix)
for row in range(size+1):
    for col in range(size+1):
        if (row, col) in moved:
            skip_count += 1
            continue
        origin = (row, col)
        value = matrix[row][col]
        for _ in range(4):
            row, col, value = move(matrix, row, col, value, size)
            op_count += 1
            mprint(matrix)
            moved.add((row, col))

print(f"{op_count=}")
print(f"{skip_count=}")
```
